```python
def step(w, b, x):
  return 1 if w * x + b > 0 else 0
```

```python
X = [i for i in range(1, 101)]
Y = [0 if x < 50 else 1 for x in X]
```

```python
def perceptron():
  w, b = 0.0, 0.0
  eta = 1e-5
  epochs = 5000

  for epoch in range(epochs):
    for x, y in zip(X, Y):
      y_pred = step(w, b, x)

      error = y - y_pred

      w += eta * error * x
      b += eta * error

  return w, b
```

```python
w, b = perceptron()
```

```python
def accuracy(w, b):
    correct = 0
    for x, y in zip(X, Y):
        if step(w, b, x) == y:
            correct += 1
    return correct / len(X) * 100

accuracy(w, b)
```

```python
import numpy as np

image = np.array([
    [1,2,3,4,5],
    [4,5,6,7,8],
    [7,8,9,1,2],
    [1,3,5,7,9],
    [2,4,6,8,0]
], dtype=float)
```

```python
filter = np.array([
    [1,0,-1],
    [1,0,-1],
    [1,0,-1]
], dtype=float)
```

```python
def iterate_regions(image, filter_size=3):
    h, w = image.shape
    f = filter_size

    for i in range(h - f + 1):
        for j in range(w - f + 1):
            region = image[i:i+f, j:j+f]
            yield i, j, region
```

```python
for i, j, region in iterate_regions(image):
    print("Position:", i, j)
    print(region)
    print()
```

```python
output = np.zeros((3, 3))

for i, j, region in iterate_regions(image):
    output[i, j] = np.sum(region * filter)

output
```

# Multi-Channel Case

```python
image = np.random.randint(0, 10, (5,5,3))
filter = np.random.randint(-2, 3, (3,3,3))

def iterate_regions(image, f=3):
    H, W, C = image.shape
    for i in range(H - f + 1):
        for j in range(W - f + 1):
            yield i, j, image[i:i+f, j:j+f, :]

for i, j, region in iterate_regions(image):
    print("pos:", i, j)
    print(region)
    print()
```

```python
output = np.zeros((3,3))

for i, j, region in iterate_regions(image):
    output[i, j] = np.sum(region * filter)

output
```

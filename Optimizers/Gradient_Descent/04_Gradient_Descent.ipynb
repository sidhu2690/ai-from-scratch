# Stochastic Gradient Descent

```python
m, c = 2.0, 3.0

X = [i for i in range(1, 100)]
Y = [m * i + c for i in X]
```

```python
def fx(w, b, x):
  return w * x + b

def gradient_w(w, b, x, y):
  return (fx(w, b, x) - y) * x

def gradient_b(w, b, x, y):
  return (fx(w, b, x) - y)
```

```python
def stochastic_gd():
  w, b = 0.0, 0.0
  eta = 1e-6
  epochs = 200

  for epoch in range(epochs):
    for x, y in zip(X, Y):
      dw = gradient_w(w, b, x, y)
      db = gradient_b(w, b, x, y)

      w -= eta * dw
      b -= eta * db
  return w, b
```

```python
w_pred, b_pred = stochastic_gd()
```

```python
y_pred = [w_pred * i + b_pred for i in X]
```

```python
def mse(y_pred, Y):
  mse_vals = [(y_pred[i] - Y[i])**2 for i in range(len(y_pred))]
  return sum(mse_vals) / len(Y)
```

```python
mse(y_pred, Y)
```

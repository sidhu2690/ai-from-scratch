```python
m, c = 2, 3
```

```python
X = [i for i in range(1, 100)]
Y = [m * i + c for i in X]
```

```python
def fx(x, w, b):
  return w*x+b

def gradient_w(w, b, x, y):
  return (fx(x, w, b) - y) * x

def gradient_b(w, b, x, y):
  return (fx(x, w, b) - y)
```

```python
def train():
  w, b = 0.0, 0.0
  eta = 0.0001
  epochs = 1000

  n = len(X)

  for epoch in range(epochs):
    dw, db = 0.0, 0.0

    for x, y in zip(X, Y):
      dw += gradient_w(w, b, x, y)
      db += gradient_b(w, b, x, y)

    dw /= n
    db /= n

    w -= eta * dw
    b -= eta * db

  return w, b
```

```python
w_pred, b_pred = train()
```

```python
y_pred = [w_pred * i + b_pred for i in X]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, Y)
plt.scatter(X, y_pred)
plt.show()
```
